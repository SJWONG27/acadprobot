```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_ollama.llms import OllamaLLM

template = """Question: {question}

Answer: Let's think step by step."""

prompt = ChatPromptTemplate.from_template(template)

model = OllamaLLM(model="llama3.2")

chain = prompt | model

chain.invoke({"question": "What is LangChain?"})
```

```python
import requests
from bs4 import BeautifulSoup

# 1. 抓取网页
url = "https://umexpert.um.edu.my/expert_search.php?search_front=C&cat=faculty"
response = requests.get(url)

# 2. 用 BeautifulSoup 解析网页
soup = BeautifulSoup(response.text, "html.parser")

# 3. 提取需要的内容
title = soup.title.string
print(f"网页标题：{title}")

# 4. 提取正文里的文字
divs = soup.find_all('div')
texts = [div.get_text(strip=True) for div in divs if div.get_text(strip=True)]

# 5. 拼接所有文字
full_text = f"网页标题：{title}\n\n" + "\n".join(texts)

# 6. 保存到 txt 文件
with open("umexpert_output.txt", "w", encoding="utf-8") as f:
    f.write(full_text)

print("✅ 已保存为 umexpert_output.txt")
```
