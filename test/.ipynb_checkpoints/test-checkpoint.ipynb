```python
import requests
from bs4 import BeautifulSoup

# 1. æŠ“å–ç½‘é¡µ
url = "https://umexpert.um.edu.my/expert_search.php?search_front=C&cat=faculty"
response = requests.get(url)

# 2. ç”¨ BeautifulSoup è§£æžç½‘é¡µ
soup = BeautifulSoup(response.text, "html.parser")

# 3. æå–éœ€è¦çš„å†…å®¹
title = soup.title.string
print(f"ç½‘é¡µæ ‡é¢˜ï¼š{title}")

# 4. æå–æ­£æ–‡é‡Œçš„æ–‡å­—
divs = soup.find_all('div')
texts = [div.get_text(strip=True) for div in divs if div.get_text(strip=True)]

# 5. æ‹¼æŽ¥æ‰€æœ‰æ–‡å­—
full_text = f"ç½‘é¡µæ ‡é¢˜ï¼š{title}\n\n" + "\n".join(texts)

# 6. ä¿å­˜åˆ° txt æ–‡ä»¶
with open("umexpert_output.txt", "w", encoding="utf-8") as f:
    f.write(full_text)

print("âœ… å·²ä¿å­˜ä¸º umexpert_output.txt")
```

```python
from sentence_transformers import SentenceTransformer
import requests
import numpy as np

# --- Step 1: Load the model ---
model = SentenceTransformer("all-mpnet-base-v2")

# --- Step 2: Embed user query ---
user_query = "Who is Dr. Su Moon Ting?"
query_embedding = model.encode(user_query, convert_to_numpy=True)  # if pyvector needs numpy

# def convert_to_pgvector_format(embedding: np.ndarray) -> str:
#     return "{" + ",".join([str(x) for x in embedding.tolist()]) + "}"

# pgvector_embedding = convert_to_pgvector_format(query_embedding)


response = supabase.rpc("match_embeddings", {
    "query_embedding": query_embedding.tolist(),
    "match_count": 3
}).execute()

top_chunks = [item["content"] for item in response.data]
context = "\n\n".join(top_chunks)

print(context)

# --- Step 5: Construct prompt ---
prompt = f"""You are a helpful assistant. Use the following context to answer the question:

Context:
{context}

Question:
{user_query}

Answer:"""

# --- Step 6: Send to Ollama ---
response = requests.post("http://localhost:11434/api/generate", json={
    "model": "llama3.2",
    "prompt": prompt,
    "stream": False
})

# --- Step 7: Output response ---
print("\nðŸ“˜ Answer from Ollama:\n")
print(response.json()['response'])
```
